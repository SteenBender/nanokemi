### Opgave 1) Ligevægte – 24 point
Der kan opløses 0,584 g bly(II)klorid i 100 mL vand ved 298K.

$$PbCl_2(s) \rightleftharpoons Pb^{2+}(aq) + 2Cl^-(aq)$$

#### a) Beregn opløselighedsproduktet af bly(II)klorid. 

```python
pb2cl = 278.10  # g/mol
g_pb2cl = 0.584  # g
mol_pb2cl = g_pb2cl / pb2cl
c_pb2cl = mol_pb2cl / 0.1  # mol/L
c_cl = 2 * c_pb2cl

Ksp = c_pb2cl * c_cl**2
text = f"""
$[Pb^{+2}]$ = {c_pb2cl:.2f} mol/L

$[Cl^-]$ is {c_cl:.2f} mol/L

The Ksp of $Pb(II)Cl_2$ is {Ksp:.2e}

___
"""
md(text)
```

#### b) Beregn opløseligheden af bly(II)klorid i en opløsning af 0,0010 M bly(II)nitrat.  

___

<center>

|Stof|$PbCL_2$|$Pb^{+2}$|$Cl^-$|
|---|---|---|---|
|start|$\infty$|0.001 M|0|
|ændring|-x|+x|+2x|
|slut| $\infty$ - x|0.001 + x|2x|

</center>

$$K_{sp} = [Pb^{+2}][Cl^-]^2 = 3.7e-5 = (0.001 + x)(2x)^2 = 4x^3 + 0.004x^2$$
$$4x^3 + 0.004x^2 - 3.7*10^{-5} = 0$$


```python
x = sp.symbols("x")
eq = 4 * x**3 + 0.004 * x**2 - Ksp
solution = sp.solve(eq, x)
real_solution = [i if i.is_real else None for i in solution]

text = f"""
<font size="4">
<center>
x (Opløsligheden) = {real_solution[0]:.2e} mol/L

___
"""
md(text)
```

### Opgave 3) Termodynamik og kinetik - 30 point
Betraft ligevægten:

$$2I(g) + Ar(g) \rightleftharpoons I_2(g) + Ar(g)$$

#### a) Udregn ΔH og ΔS for reaktionen ud fra tabelværdierne

```python
dh_I = 106.76  # kJ/mol
dh_Ar = 0  # kJ/mol
dh_I2 = 62.42  # kJ/mol

ds_I = 180.79  # J/mol*K
ds_Ar = 0  # J/mol*K
ds_I2 = 260.69  # J/mol*K


delta_H = dh_I2 - 2 * dh_I
delta_S = ds_I2 - 2 * ds_I

text = f"""
___

$\\Delta H^\\circ = \\Delta H_{{I_2}} - 2 \\Delta H_I = {dh_I2} - 2 \\cdot {dh_I} = {delta_H:.2f}$ kJ/mol

$\\Delta S^\\circ = \\Delta S_{{I_2}} - 2 \\Delta S_I = {ds_I2} - 2 \\cdot {ds_I} = {delta_S:.2f}$ J/mol*K

___
"""
md(text)
```

#### b) Opskriv funktionen for ΔG(T) for reaktionen og plot ΔG(T) fra T = 0K til T = 400K. Der må bruges én graf i besvarelsen.

```python
def dg(T):
    return delta_H - T * delta_S / 1000


t_range = np.arange(401)

g_vals = dg(t_range)

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(t_range, g_vals, label="$\Delta G = -151.10 - T \cdot (-0.1)$")
ax.grid()
ax.set_xlabel("Temperature (K)")
ax.set_ylabel("Gibbs free energy (kJ/mol)")
ax.legend()

text = f"""
The Gibbs free energy is given by the equation:

$\\Delta G = \\Delta H - T \\Delta S$

"""

md(text)
```
