Ascorbinsyre har sumformlen C6H8O6 og er måske bedre kendt som C-vitamin. Det er en divalent

#### a) Skriv reaktionsligninger op for den første deprotonisering af ascorbinsyre ved brug af sumformlen og ved brug af de generiske HA.
___

$HA \rightleftharpoons H^+ + A^-$  
$C_6H_8O_6 \rightleftharpoons H^+ + C_6H_7O_6^-$
___

#### b) Beregn pH i en vandig opløsning af 0,400 g ascorbinsyre i 0,100 L.
___

```python
m = 0.4  # g
vol = 0.1  # L
mol_mass = 176.12  # g/mol

mol = m / mol_mass
c = mol / vol
print(f"Moles: {mol:.3e} mol")
print(f"Concentration: {c:.3e} mol/L")
```

Reaktions tabel:
|Start| $C_6H_8O_6$ | $H^+$ | $C_6H_7O_6^-$ |
|---|---|---|---|
|Ændring| -x | +x | +x |
|Slut| $0.0227 - x$ | $1.0 \times 10^{-7} + x$ | $x$ |

Note: Vand har en ionsproduks:
$K_w = [H^+][OH^-] = 1.0 \times 10^{-14} \rightarrow  [H^+] = [OH^-] = 1.0 \times 10^{-7}$


$K_a = \frac{[H^+][C_6H_7O_6^-]}{[C_6H_8O_6]} = 4.27 \times 10^{-5}$

$4.27 \times 10^{-5} = \frac{(1.0 \times 10^{-7} + x)x}{0.0227 - x}$

WolframAlpha solution:
$x = -0.001 \: \text{or} \: 0.00096$


$pH = -\log(0.00096) = 3.02$

____


#### c) Argumenter for at det er OK du kun har regnet på første deprotonering i b).
____

$pKa_1 = -log(4.27 \times 10^{-5}) = 4.37$  
$pKa_2 = -log(4.57 \times 10^{-12}) = 11.34$

En pH 3.02 er lavere end $pKa_1$ og og $pKa_2$ er over double så høj som $pKa_1$. Derfor er det OK at kun regne på første deprotonering.

#### a) Plot de to dataserier og vis om reaktionerne er af 0., 1. eller 2. orden. Der må bruges seks grafer i besvarelsen.

### Opgave 4 - Elektrokemi
Ascorbinsyre (C6H8O6) kan oxideres til dehydroascorbinsyre (C6H6O6), som igen kan reduceres til
ascorbinsyre. Når ascorbinsyre oxideres, kan mange andre metaller reduceres:

#### a) Beregn den elektromotoriske kraft i elementet, når potentialerne for reduktion og oxidation er Ered =+1,23 V og Ered =+0,472 V:
___
Pt(s)| C6H6O6 (aq, 0,15 M), C6H8O6 (aq, 0,10 M) HCl (aq, 0,11 M)|| O2 (g, 0,20 bar)| HCl (aq, 0,11 M)|Pt(s)


```python
R = 8.314
T = 298
F = 96485
q = 0.1**2 * 0.2 / 0.15**2

E_cell = 1.23 - 0.472
rt_nf_const = R * T / (4 * F)
E = E_cell - (rt_nf_const * np.log(q))
print(f"E = {E:.3f} V")
```

___
#### b) Svovl kan indgå i en redox reaktion med dehydroascorbinsyre, ogskriv ligevægten og beregn ligevægtskonstanten for reaktionen mellem dehydroascorbinsyre og hydrogensulfid.

___
$C_6H_6O_6 + H_2S \rightleftharpoons C_6H_8O_6 + S + H_2O$

$H_2S \rightarrow 2H^+ + S : 0.142V$ Oxidation

```python
E_celle = 0.472 - 0.142
G = -2 * F * E_celle
k = np.exp(-G / (R * T))
print(f"E_celle = {E_celle:.3f} V")
print(f"G = {G:.3e} J")
print(f"k = {k:.3e}")
```
