#### c) Hvor stort et volumen 10 M saltsyre skal der til at opløse 0,1 mmol sølv(I)klorid

$$ K_{sp} = [Ag^+][Cl^-] = 1.0 \cdot 10^{-10}$$
$$ Ag^+ + 4Cl^- \rightleftharpoons [Ag(Cl)_4]^{3-}$$

#### d) Opskriv ligevægten for reaktionen fra sølv(I)cyanid til dicyanoargentat(I)

___
$$AgCN = Ag^+ + CN^-$$
$$ K_{sp} = [Ag^+][CN^-] = 1.2 \cdot 10^{-16}$$
$$ Ag^+ + 2CN^- \rightleftharpoons [Ag(CN)_2]^-$$
$$ k_{kompleks} = \frac{[Ag(CN)_2]^-}{[Ag^+][CN^-]^2} = 1.0 \cdot 10^{20}M^{-2}$$
$$ AgCN + CN^- \rightleftharpoons [Ag(CN)_2]^-$$
___

#### e) Beregn ligevægtskonstanten og ΔG for reaktionen

___
$$ K = \frac{[Ag(CN)_2]^-}{[AgCN][CN^-]} = [Ag^+] [CN]\frac{[Ag(CN)_2]}{[Ag^+][CN^-]^2} = K_{sp} \cdot k_{kompleks} = 1.2 \cdot 10^{-16} \cdot 1.0 \cdot 10^{20} = 1.2 \cdot 10^4$$

$$ \Delta G = -RTln(K) = -8.314 \cdot 298 \cdot ln(1.2 \cdot 10^4) = -8.314 \cdot 298 \cdot 9.39 = -23.4 kJ/mol$$

#### Opgave 3. Termodynamik og kinetik – 30 point
I har denne ligevægt, der indstiller sig når N2O5 opløses i brom:

$$2N_2O_5 + Br_2(l) \rightleftharpoons 4NO_2 + O_2  + Br_2(l)$$

#### a) Udregn ΔH og ΔS for reaktionen ud fra tabelværdierne.

___
$$ 4NO_2 + O_2  + Br_2(l) \rightarrow 2N_2O_5 + Br_2(l)$$

$$\Delta H = \Sigma \Delta H_{f,produkter} - \Sigma \Delta H_{f,reaktanter} = 4 \cdot 33.2 + 0 - 2 \cdot 11.3 = 108.4 kJ/mol$$

$$\Delta S = \Sigma \Delta S_{f,produkter} - \Sigma \Delta S_{f,reaktanter} = 4 \cdot 240.1 + 205 - 2 \cdot 346.55 = 472.1 J/molK$$
___

#### b) Opskriv funktionen for ΔG(T) og beregn ligevægtskonstanten ved 298K 

___
$$\Delta G = \Delta H - T\Delta S$$
$$\Delta G = 108400 - 298 \cdot 472.1 = -32.28 kJ/mol$$
$$K(298) = e^{\frac{-\Delta G}{RT}} = e^{\frac{-32280}{8.314 \cdot 298}} = 539.000 M^3$$
___

#### d) Brug ovenstående kinetiske data til at bestemme reaktionsordenen og  hastighedskonstanten. Der må bruges tre grafer i besvarelsen

```python
t = np.array([0, 200, 400, 600, 1000])
c = np.array([0.11, 0.073, 0.048, 0.032, 0.014])
log_c = np.log(c)
one_over_t = 1 / t

lon_0_fit = np.polyfit(t, c, 1)
lin_1_fit = np.polyfit(t, log_c, 1)


fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].plot(t, c, "o")
ax[0].plot(t, np.polyval(lon_0_fit, t))
ax[0].set(xlabel="Time [s]", ylabel="Concentration [M]")
ax[1].plot(t, log_c, "o")
ax[1].plot(t, np.polyval(lin_1_fit, t))
ax[1].set(xlabel="Time [s]", ylabel="ln(Concentration) [M]")
ax[2].plot(one_over_t, log_c, "o")
ax[2].set(xlabel="1/Time [1/s]", ylabel="ln(Concentration) [M]")

text = f" $c(t) = {lin_1_fit[0]:.2e} \cdot t {lin_1_fit[1]:.2f}$"
ax[1].text(0.2, 0.9, text, transform=ax[1].transAxes, fontsize=14)
plt.tight_layout()
```
